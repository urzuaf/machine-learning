## Ч 2. Preprocesamiento

Antes de entrenar un modelo de Machine Learning, es necesario **preparar adecuadamente los datos**. Este proceso se conoce como *preprocesamiento* y consiste en dejar el conjunto de datos en un formato que los algoritmos puedan utilizar de manera efectiva.

### Pasos a seguir:

1. **Eliminar columnas innecesarias**: Algunas columnas como `PassengerId`, `Name`, `Ticket` o `Cabin` no aportan informaci贸n 煤til al modelo o contienen datos dif铆ciles de procesar autom谩ticamente (como texto libre o demasiados valores 煤nicos).

2. **Manejo de valores faltantes**:
   - La columna `Age` tiene valores faltantes que puedes reemplazar por la **mediana**, ya que es menos sensible a valores extremos.
   - La columna `Embarked` tiene pocos valores faltantes, as铆 que puedes usar la **moda** (valor m谩s frecuente) para completarlos.

3. **Codificaci贸n de variables categ贸ricas**:
   - Los algoritmos no pueden trabajar directamente con texto. Por eso, aplicamos **One Hot Encoding** a las variables categ贸ricas como `Sex` y `Embarked`, transform谩ndolas en columnas binarias (0 y 1).
   - Usar `drop_first=True` evita multicolinealidad y reduce la cantidad de columnas.

> 锔 **Importante:** El preprocesamiento tiene un gran impacto en la calidad del modelo. Un paso mal hecho puede afectar seriamente el desempe帽o del algoritmo.

Al finalizar esta etapa, deber铆as tener un dataset **limpio, sin valores nulos y completamente num茅rico**.

```python
clean_df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
clean_df.fillna({'Age': clean_df['Age'].median()}, inplace=True)
clean_df.fillna({'Embarked': clean_df['Embarked'].mode()}, inplace=True)
clean_df = pd.get_dummies(clean_df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python

clean_df.head()
```

##  3. Entrenamiento del modelo de rbol de Decisi贸n

En esta etapa, construiremos un modelo predictivo utilizando un **rbol de Decisi贸n**, una t茅cnica de clasificaci贸n que divide los datos en ramas sucesivas para tomar decisiones basadas en las caracter铆sticas del conjunto de datos.

### Pasos a seguir:

1. **Separar las variables predictoras (X) y la variable objetivo (y)**.  
   En este caso, `Survived` ser谩 nuestra variable objetivo, y el resto de las columnas las usaremos como predictores.

2. **Dividir los datos en conjuntos de entrenamiento y prueba**.  
   Esto se hace para evaluar el modelo con datos que no ha visto antes. Usa `train_test_split` para crear una divisi贸n, por ejemplo 70% entrenamiento y 30% prueba.

3. **Crear y entrenar el modelo con `DecisionTreeClassifier`**.  
   Puedes usar el par谩metro `max_depth` para limitar la profundidad del 谩rbol y evitar sobreajuste.

Cuando termines esta secci贸n, tu modelo estar谩 entrenado y listo para ser evaluado y visualizado.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = clean_df.drop(columns=["Survived"])
y = clean_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier(max_depth=5, random_state=42)

model.fit(X_train, y_train)


```

##  4. Visualizaci贸n del 谩rbol

Una de las grandes ventajas de los 谩rboles de decisi贸n es que su estructura se puede visualizar de forma intuitiva, lo que facilita la **interpretaci贸n del modelo**.

En esta secci贸n usaremos la funci贸n `plot_tree()` de `sklearn` para graficar el 谩rbol entrenado. En el gr谩fico:

- Cada **nodo interno** representa una condici贸n de decisi贸n sobre alguna variable.
- Cada **rama** muestra el resultado de esa decisi贸n (verdadero o falso).
- Cada **nodo terminal** (o hoja) muestra la clase predicha y cu谩ntas muestras llegaron a ese punto.

### Sugerencia:
Puedes ajustar los par谩metros de `plot_tree()` como `feature_names`, `class_names` y `filled=True` para mejorar la legibilidad del 谩rbol. Aseg煤rate de usar `plt.figure(figsize=(ancho, alto))` para que la visualizaci贸n no se corte.

**Objetivo:** Entender c贸mo el modelo est谩 tomando decisiones, qu茅 variables usa y con qu茅 prioridad.

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))  

plot_tree(
    model,
    feature_names=X.columns,      
    class_names=["Not survived", "Survived"],  
    filled=True,                   
    rounded=True,                  
    fontsize=10                    
)

plt.title("rbol de decisi贸n predicci贸n supervivencia titanic")
plt.show()
```

##  5. Evaluaci贸n del modelo

Una vez entrenado el modelo, es fundamental **evaluar su desempe帽o** para entender qu茅 tan bien est谩 funcionando. En esta secci贸n usaremos datos de prueba (no vistos por el modelo durante el entrenamiento) para calcular m茅tricas que nos ayuden a juzgar la calidad de las predicciones.

### M茅tricas que utilizaremos:

- **Matriz de confusi贸n (`confusion_matrix`)**: muestra la cantidad de verdaderos positivos, falsos positivos, verdaderos negativos y falsos negativos. Es 煤til para entender errores espec铆ficos del modelo.
- **Reporte de clasificaci贸n (`classification_report`)**: incluye precisi贸n (*precision*), exhaustividad (*recall*), y la m茅trica F1 para cada clase.
- **Exactitud (`accuracy_score`)**: indica qu茅 porcentaje total de predicciones fue correcto.

Estas m茅tricas te permitir谩n **identificar si el modelo est谩 desbalanceado**, si tiende a predecir m谩s una clase que otra, y si necesita ajustes en par谩metros o preprocesamiento.

>  Recuerda: un modelo con buena precisi贸n pero bajo *recall* puede ser problem谩tico si nos interesa no dejar pasar casos positivos (como sobrevivientes).

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(classification_report(y_test, y_pred, target_names=["Not survived", "Survived"]))
```

```python
print(f"{accuracy_score(y_test, y_pred):.2f} ")
```
